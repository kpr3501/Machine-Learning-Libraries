```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

The main differnce betweeen the rug plot and dist plot is dist plot presents the data in the form of histograms but rug plot presents the data in the form of lines on the same level
suppose    
1-In the dist plot the data between 10 to 20 at x axis is 45 at y axis
2-but in the rug plot  the data between 10 to 20  is 45 lines betweeen 10 to 20 at x axis

#  Categorical plot

## 1-Bar plot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='size',y='total_bill',data=tips)
```

estimator arguments tells that on what basis we plot the graph bydefault it is mean of the data but we can change as what we want to present like standard deviation of the data or variance of the data

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) #np.std means standard ddeviation of the data
```
