# Matplotlib

# functional plot

```python
plt.plot(x,y,'b-')
#plt.show() it isused when we are not working on jupyter notebook .It is udsed to show the plot 
# if want to change color of the curve we use "color first letter and -"
```

```python
plt.plot(x,y,'r-')
plt.xlabel("X  label")
plt.ylabel("Y label")
plt.title("title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.xlabel("X  label")
plt.ylabel("Y label")
plt.title("title")

#subplot means create a subplot in a single plot
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.xlabel("X  label")
plt.ylabel("Y label")
plt.title("title")
```

# object oriented plot

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])#we always pass list when we add axes
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("title")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.set_xlabel('x label')
axes1.set_ylabel('Y label')
axes1.set_title(' Larger  Plot')

axes2=fig.add_axes([0.2,0.5,0.4,0.3])#these are the size from the left ,right ,top,bottom
axes2.set_xlabel('x label')
axes2.set_ylabel('Y label')
axes2.set_title('smaller plot')
# Now plot these figure
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=4)
#plt.tight_layout()#fixing the no. of the plots
```

```python
fig,axes=plt.subplots(nrows=2,ncols=4)
fig.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
#(for current_ax in axes:
#    current_ax.plot(x,y))
#we can use for loop also to plot on the figure
axes[0].plot(x,y)
axes[1].plot(y,x)
```

# Figure size and DPI

```python
fig=plt.figure(figsize=(8,2),dpi=100)# we pass the sizee of tuuple as a tuple which defined that size that  size of the box like 3 by 2
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(figsize=(8,2))
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
fig.tight_layout()
```

# add "legend" to identify the graphs


```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X square")
ax.plot(x,x**3,label="x cube ")
ax.legend()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X square")
ax.plot(x,x**3,label="x cube ")
ax.legend(loc=0) #best location is "0"
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X square")
ax.plot(x,x**3,label="x cube ")
ax.legend(loc=1)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X square")
ax.plot(x,x**3,label="x cube ")
ax.legend(loc=10)# "10 " is used for center 1 to 10 n is used for ploting legend
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X square")
ax.plot(x,x**3,label="x cube ")
ax.legend(loc=(0.1,0.1)) #we can also pass tuple for defining the box location  of legend
```

# Control over axis appearance

# Plot range 

we can figure the ranges of the axes using the "set_ylim" and "set_xlim" methods in the axis object, or axis ("tight") for automatically getting "tightly fitted" axes ranges:

```python
 fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
plt.plot(x,y,color="red",lw=2,ls="--")
ax.set_xlim([0,1])#we pass the range of x axis upto which we want to show the figure
```

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
plt.plot(x,y,color="red",lw=2,ls="--")
ax.set_xlim([0,2])
```

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
plt.plot(x,y,color="red",lw=2,ls="--")
ax.set_ylim([0,1])
```

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
plt.plot(x,y,color="red",lw=2,ls="--")
ax.set_ylim([0,2])
```

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
plt.plot(x,y,color="red",lw=2,ls="--")
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
